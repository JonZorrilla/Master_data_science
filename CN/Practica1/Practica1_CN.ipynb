2. ¬øSe te ocurre una manera en la cual el c√°lculo de los valores singulares de una matriz permita calcular un polinomio de regresi√≥n adecuado?

S√≠. Si descomponemos un polinomio de regresi√≥n en una matriz de componentes A y un vector correspondiente a las variables x, el resultado vendr√° dado por la ecuaci√≥n $Ax = y$. Esta matriz A puede descomponerse en una matriz diagonal donde cada elemento de la diagonal corresponde a un autovalor de la siguiente manera: $ A = U S V'$. Entonces, una vez se resuelve la ecuaci√≥n singular y se hallan los autovalores y autovectores, el c√°lculo se reduce a $U S V' x = y$, y se podr√≠a hallar el resultado $x$ haciendo uso de esta ecuaci√≥n.


3. ¬øComo podr√≠as transformar el problema de encontrar un polinomio de regresi√≥n en un problema de producto de matrices?

Usaremos las matrices que hemos definido en el apartado anterior para definir la siguiente ecuaci√≥n en n dimensiones: $y_i = \beta_0 + \beta_1 x_1 + ... + \beta_n x_n + \epsilon_i$.

$ Y = \begin{pmatrix} y_0 \\ y_1 \\ ... \\ y_n \end{pmatrix} $

$ X = \begin{pmatrix} 1 & x_1 & x_1^2 & ... & x_1^n \\ 1 & x_1 & x_1^2 & ... & x_1^n \\ ... \\ 1 & x_n & x_2^2 & ... & x_n^n \end{pmatrix} $

$ \beta = \begin{pmatrix} \beta_0 \\ \beta_1 \\ ... \\ \beta_n \end{pmatrix} $

$ \epsilon = \begin{pmatrix} \epsilon_0 \\ \epsilon_1 \\ ... \\ \epsilon_n \end{pmatrix} $

De esta manera, $ Y = X\beta + \epsilon$ es la operaci√≥n matricial a resolver. Descomponiendo la matrix $X$ como $X = P^T D P$. Realizando los siguiente c√°lculos:
\begin{equation}
X'X = V S^2 V'
\end{equation}
Entonces, si conocemos $S$ y $V$, podemos conoces $(X'X)^{-1}$.

Considere el siguiente conjunto de datos:



vamos a buscar el polinomio $P_2(x) = a_0 + a_1x + a_2x^2$ que minimiza el error al aproximar los datos anteriores usando la norma est√°ndar de Rn. Es decir, vamos a usar una regresi√≥n cuadr√°tica.

a) Vamos a escribir el problema de regresi√≥n como un problema de interpolaci√≥n sobredimensionado asumiendo que el polinomio P2(x) pasa por todos los puntos $x_i$,$y_i$. Para ello escribiremos el problema en la forma $Xa = y$ donde $X$ es una matriz 5 x 3, $a$ es un vector 3 x 1 e $y$ es un vector 5x1. 

```python
# Hacmemos uso de la matriz de Vandemonde
x = np.array([0.0, 0.25, 0.5, 0.75, 1.0])
y = np.array([1, 1.2840, 1.6487, 2.1170, 2.7183]).reshape(5,1)
X = np.vander(x, N=3, increasing=True)
print(X)
```

```python
# Descomponemos esta matriz X en autovalores y autovectores.
U, s, V = np.linalg.svd(X, full_matrices=True)
print('U:\n',U)
print('S:\n',s)
print('V:\n',V)
```

```python
# Hallamos la matriz asociada S.
S = np.zeros((5,3))
np.fill_diagonal(S, s)
print('S:\n',S)
```

```python
# Multiplicamos sobre U y normalizamos para hallar los vectores:
C = U.T@y
c = C[:3].reshape(3,1)
z = (c.T/s[:3]).reshape(3,1)
print('c:\n', c)
print('z:\n', z)
```
